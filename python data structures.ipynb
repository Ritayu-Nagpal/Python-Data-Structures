### Stack Implementation

```python
class Stack():
    def __init__(self):
        self.stack = []
        self.size = 0
    
    def push(self, inpt_val):
        self.stack.append(inpt_val)
        self.size+=1
        return self.stack
    
    def pop(self):
        if self.size == 0:
            return "Empty Stack"
        pop_val = self.stack[self.size-1]
        del self.stack[self.size-1]
        self.size-=1
        return pop_val
    
    def peek(self):
        if self.size == 0:
            return "Empty Stack"
        else:
            return self.stack[self.size-1]
    
    def IsEmpty(self):
        if self.size == 0:
            return True
        else:
            return False
    
    def get_size(self):
        return self.size
    
    def find(self,val):
        if val in self.stack:
            return True
        else:
            return False
        
# Creating my stack:

my_stack = Stack()
my_stack.push(5)
my_stack.push(11)
my_stack.pop()
my_stack.push(65)
my_stack.push(49)
my_stack.get_size()
my_stack.IsEmpty()
my_stack.find(11)
my_stack.find(65)
my_stack.peek()
```

### Queue Implementation

```python
class Queue():
    def __init__(self):
        self.queue = []
        self.size = 0
    
    def enqueue(self, val):
        self.queue.append(val)
        self.size+=1
        return self.queue
    
    def dequeue(self):
        if self.size == 0:
            return "Empty Queue"
        dequeue_val = self.queue[0]
        del self.queue[0]
        self.size-=1
        return dequeue_val
    
    def top(self):
        if self.size == 0:
            return "Empty Queue"
        return self.queue[0]
    
    def get_size(self):
        return self.size
    
    def IsEmpty(self):
        if self.size == 0:
            return True
        return False
    
    def find(self, val):
        if val in self.queue:
            return True
        return False
    
# creating my queue

my_q = Queue()
my_q.enqueue(5)
my_q.enqueue(11)
my_q.enqueue(17)
my_q.dequeue()
my_q.top()
my_q.IsEmpty()
my_q.get_size()
my_q.find(12)
```

### Linked List Implementation

```python
class Node():
    def __init__(self, val, n=None, p=None):
        self.value = val
        self.next = n
        self.previous = p

class Linked_List():
    def __init__(self, r=None):
        self.root = r
        if r is None:
            self.size = 0
        else:
            self.size = r.size
        
    def add(self, val):
        nxt_node = Node(val)
        if self.size == 0:
            self.root = nxt_node
            self.size+=1
        else:
            nxt_node.next = self.root
            self.root = nxt_node
            self.size+= 1
            
    def find(self, f_val):
        if self.size == 0:
            return "Empty Linked List"
        curr_node = self.root
        while curr_node is not None:
            if curr_node.value == f_val:
                return True
            curr_node = curr_node.next
        return False
    
    def remove(self, r_val):
        if self.size == 0:
            return "Empty Linked List"
        elif self.size == 1:
            if self.root.value == r_val:
                self.root = None
                self.size-=1
                return True
            return "Value not in list"  
        else:
            if self.root.value == r_val:
                self.root = self.root.next
                self.size-=1
            
            curr_node = self.root
            counter = 0
            while curr_node.next is not None:
                nxt_node = curr_node.next
                if nxt_node.value == r_val:
                    if nxt_node.next is None:
                        curr_node.next = nxt_node.next
                        self.size-=1
                        return True
                    else:
                        curr_node.next = nxt_node.next
                        self.size-=1
                        counter+=1
                curr_node = curr_node.next
            if counter>0:
                return True
        return "Value not in list"
    
    def print_list(self):
        if self.size == 0:
            return "Empty List"
        elif self.size == 1:
            return self.root
        else:
            lst = []
            curr_node = self.root
            while curr_node is not None:
                lst.append(curr_node.value)
                curr_node = curr_node.next
            return lst

# creating my link list
my_ll = Linked_List()
my_ll.add(5)
my_ll.add(8)
my_ll.add(12)
my_ll.add(8)
my_ll.add(11)
my_ll.remove(12)
my_ll.find(8)
my_ll.print_list()
```

### Doubly Linked List Implementation

```python
class Node():
    def __init__(self, val, n=None, p=None):
        self.value = val
        self.next = n
        self.previous = p

class D_Linked_List():
    def __init__(self, r=None):
        self.root = r
        if r is None:
            self.size = 0
        else:
            self.size = r.size
            self.last = r.last
        
    def add(self, val):
        nxt_node = Node(val)
        if self.size == 0:
            self.root = nxt_node
            self.last = nxt_node
            self.size+=1
        else:
            nxt_node.next = self.root
            self.root.previous = nxt_node
            self.root = nxt_node
            self.size+= 1
            
    def find(self, f_val):
        if self.size == 0:
            return "Empty Linked List"
        curr_node = self.root
        while curr_node is not None:
            if curr_node.value == f_val:
                return True
            curr_node = curr_node.next
        return False
    
    def remove(self, r_val):
        if self.size == 0:
            return "Empty Linked List"
        elif self.size == 1:
            if self.root.value == r_val:
                self.root = None
                self.size-=1
                return True
            return "Value not in list"  
        else:
            if self.root.value == r_val:
                self.root = self.root.next
                self.root.previous = None
                self.size-=1
            
            curr_node = self.root
            counter = 0
            while curr_node.next is not None:
                nxt_node = curr_node.next
                nxt2_node = nxt_node.next
                if nxt_node.value == r_val:
                    if nxt_node.next is None:
                        curr_node.next = nxt_node.next
                        nxt_node.previous = None
                        self.size-=1
                        return True
                    else:
                        curr_node.next = nxt_node.next
                        nxt2_node.previous = nxt_node.previous
                        self.size-=1
                        counter+=1
                curr_node = curr_node.next
            if counter>0:
                return True
        return "Value not in list"
    
    def print_list(self):
        if self.size == 0:
            return "Empty List"
        elif self.size == 1:
            return self.root
        else:
            lst = []
            curr_node = self.root
            while curr_node is not None:
                lst.append(curr_node.value)
                curr_node = curr_node.next
            return lst

# creating my link list
my_dll = D_Linked_List()
my_dll.add(5)
my_dll.add(8)
my_dll.add(12)
my_dll.add(8)
my_dll.add(11)
my_dll.remove(5)
my_dll.find(11)
my_dll.print_list()
```
